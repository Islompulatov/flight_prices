```python
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python

# Urls with different cities
# https://www.edreams.net/travel/#results/type=O;dep=2022-10-05;from=WAW;to=BCN;internalSearch=true
# https://www.edreams.net/travel/#results/type=O;dep=2022-10-06;from=WAW;to=BCN;internalSearch=true
# https://www.edreams.net/travel/#results/type=O;dep=2022-10-07;from=WAW;to=BCN;internalSearch=true
cities = ['WAW', 'BCN', 'AMS', 'PRG', 'PAR', 'MCM', 'ROM', 'BER', 'OSL']
list_prices = []
air_names = []
dep_city = []
arr_city = []
dep_time= []
arr_time = []
stops = []
duration_of_flight = []
add_info = []
for i in cities:
    url = f"https://www.kayak.com/flights/LON-{i}/2022-10-12?sort=bestflight_a"
            
    pth = "chromedriver.exe"
    driver = webdriver.Chrome(pth)
    driver.get(url)
    sleep(30)
    # button = '//button[@class="didomi-components-button didomi-button didomi-dismiss-button didomi-components-button--color didomi-button-highlight highlight-button"]'
    # driver.find_element_by_xpath(button).click()
    info_window =driver.find_elements_by_xpath('//div[@class="inner-grid keel-grid"]')
    print(len(info_window))
    sleep(5)
    for webelement in info_window:
        elementhtml = webelement.get_attribute('outerHTML')
        soup = BeautifulSoup(elementhtml, 'html.parser')
        prices = soup.find('div', class_ = 'above-button')
        temp_prices = prices.find('span', class_ = 'price-text')
        air_name = soup.find('div', dir = 'ltr')
        cities_duration = soup.find('div', class_ = 'section duration allow-multi-modal-icons')
        arr_dep_cities = cities_duration.find('div', class_ = 'bottom')
        times = soup.find('div', class_ = 'top')
        f_stops = soup.find('span', class_ = 'stops-text')
        total_time = cities_duration.find('div', class_ =  'top')
        addit_info = soup.find('span', class_ = 'label')
        
        try:
            add_info.append(addit_info.text.strip('\n'))
        except:
            add_info.append("NaN")  
       

        list_prices.append(temp_prices.text.strip('\n$'))
        air_names.append(air_name.text.strip('\n'))
        dep_city.append(arr_dep_cities.text.strip().split()[0])
        arr_city.append(arr_dep_cities.text.strip().split()[2])
        dep_time.append(times.text.strip().split()[:2])
        arr_time.append(times.text.strip().split()[3:])
        stops.append(f_stops.text.strip('\n'))
        duration_of_flight.append(total_time.text.strip('\n'))

    
    print(list_prices)
    print(air_names) 
    print(dep_city)
    print(arr_city)
    print(dep_time)
    print(arr_time)
    print(stops)
    print(duration_of_flight) 
    print(add_info)
    
    
```

```python
lon_to_10cities_data = pd.DataFrame({'flight_prices':list_prices, 'company_names':air_names, 'dep_city':dep_city, 'arrival_city':arr_city,
 'transfer_stops':stops, 'duration_of_flight':duration_of_flight, 'dep_time':dep_time, 'arrival_time':arr_time, 'additional_info':add_info})

lon_to_10cities_data.to_csv('lon_to_10cities_data', index=False)
```
