```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('datasets/flight_data')
df
```

```python
print(df.dep_city.unique())
print(df.arrival_city.unique())
```

```python
df.replace(['WMI', 'WAW', 'AMS', '6BT', 'BCN', 'BER', 'FRA', 'STN', 'LTN',
       'LHR', 'LGW', 'LON', 'LCY', 'TRF', 'OSL', 'BVA', 'ORY', 'CDG',
       'PAR', 'PRG', 'CIA', 'FCO', 'ROM', '6BU', 'T6Q'], ['Warsaw', 'Warsaw', 'Amsterdam', 'London', 'Barcelona', 'Berlin', 'Paris', 'London', 'London',
       'London', 'London', 'London', 'London', 'Oslo', 'Oslo', 'Paris', 'Paris', 'Paris',
       'Paris', 'Prague', 'Rome', 'Rome', 'Rome', 'London', 'Oslo'], inplace=True)
df
```

```python
df.arrival_city.unique()
```

```python
df.transfer_stops.unique()
```

```python
df.transfer_stops.replace(['nonstop ', '1 stop ', '2 stops ', '2 changes ', '0 changes ',
       '1 change ', '3 changes ', '5 changes ', '4 changes ',
       '7 changes ', '3 stops '], ['non-stop', '1 stop', '2 stops', '2 stops', 'non-stop',
       '1 stop', '3 stops', '5 stops', '4 stops',
       '7 stops', '3 stops'], inplace=True)
df.transfer_stops.unique()     
```

```python
df.travel_dates.unique()
```

```python
df.replace(['2022-10-11', '2022-10-12', '2022-10-13', '2022-10-14', '2022-10-15',
       '2022-10-16', '2022-10-17', '2022-10-18', '2022-10-19'], ['11/10/2022', '12/10/2022', '13/10/2022', '14/10/2022', '15/10/2022',
       '16/10/2022', '17/10/2022', '18/10/2022', '19/10/2022'], inplace=True)
df       
```

```python
df['dep_time']=df['dep_time'].str.replace(", ", " ").str.strip("[]'").str.replace("' '", " ")
df['arrival_time']=df['arrival_time'].str.replace(", ", " ").str.strip("[]'").str.replace("' '", " ").str.replace(" +1", "", regex=False).str.replace(" +2", "", regex=False)
```

```python
df
```

```python
df['Dep_Time'] = pd.to_datetime(df['dep_time'], format='%I:%M %p').dt.strftime('%H:%M')
df['Arrival_Time'] = pd.to_datetime(df['arrival_time'], format='%I:%M %p').dt.strftime('%H:%M')
df
```

```python
df.drop(["dep_time", "arrival_time"], axis=1, inplace=True)
```
